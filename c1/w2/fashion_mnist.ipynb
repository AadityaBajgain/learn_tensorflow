```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import certifi, os
os.environ["SSL_CERT_FILE"] = certifi.where()
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
index = 0
np.set_printoptions(linewidth = 320)

print(f"Label: {train_labels[index]}")
print(f"\n IMAGE PIXEL ARRAY: \n\n {train_images[index]}")
```

```python
plt.imshow(train_images[index])
plt.colorbar()
```

```python
train_images = train_images / 255
test_images = test_images / 255
```

```python
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(28,28)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,tf.nn.relu),
    tf.keras.layers.Dense(10,tf.nn.softmax)
])
```

```python
model.compile(
    optimizer = tf.optimizers.Adam(),
    loss = "sparse_categorical_crossentropy",
    metrics = ["Accuracy"]
)
```

```python
class callback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs ={}):
        if(logs.get("Accuracy") >= .8):
            self.model.stop_training = True

callbacks = callback()
```

```python
model.fit(train_images, train_labels, epochs = 5, callbacks=[callbacks])
```

### more you train the model on the training set more it overfits and perform poorly on the newer data

```python
model.evaluate(test_images, test_labels)
```
